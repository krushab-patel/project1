# Anomaly detection

```python
data.shape
```

```python
q1 = data["Sale Price"].quantile(0.25) 
q3 = data['Sale Price'].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5*IQR) 
max_range = q3 + (1.5*IQR) 
min_range,max_range
```

```python
data = data[data['Sale Price']<=max_range]
```

```python
data.shape
```

```python
plt.figure(figsize=[5,4])
sns.boxplot(x='Sale Price',data=data)
```

```python
plt.figure(figsize=[5,4])
sns.lineplot(data=data,x="Car Year",y="Sale Price")
```
